```python
import os
import csv
import yfinance as yf

# Define a list of ticker symbols
tickers = ticker_list

# Define the filename for the CSV file
filename = "ticker_data.csv"

# Define the directory path for the CSV file
directory = "/content/drive/MyDrive/05.23 ADA FINAL PROJECT/"

# Create the directory if it does not exist
if not os.path.exists(directory):
    os.makedirs(directory)

# Join the directory path and filename to create the full file path
file_path = os.path.join(directory, filename)

# Define the fieldnames for the CSV file
fieldnames = ['ticker'] + list(yf.Ticker('AAPL').info.keys()) + ['address2', 'fax', 'state', 'lastSplitDate', 'lastSplitFactor', 'industrySymbol', 'lastDividendDate', 'lastDividendValue', 'earningsGrowth', 'earningsQuarterlyGrowth', 'dividendRate', 'dividendYield', 'fiveYearAvgDividendYield', 'exDividendDate']

# Iterate over the tickers and append each ticker's information to the CSV file
with open(file_path, mode='w', newline='') as csvfile:
    
    # Create a CSV writer object
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    
    # Write the header row to the CSV file
    writer.writeheader()
    
    for ticker in tickers:
        
        # Use the yf.Ticker() method to get the stock data
        ticker_info = yf.Ticker(ticker).info
        
        # Add the ticker symbol to the dictionary
        ticker_info['ticker'] = ticker
        
        # Write the data row to the CSV file
        writer.writerow(ticker_info)
        
        print(f"{ticker} data saved successfully")
    
print(f"All ticker data saved to {file_path}")
```
